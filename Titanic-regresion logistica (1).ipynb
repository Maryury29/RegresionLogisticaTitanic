```python
#2.Preprocesar los datos limpiándolos, tratando valores faltantes y transformándolos según sea necesario.
```

```python
Datos.info()
```

```python
# Revisión de datos Faltantes

print("Cantidad de registros con cero en la columna Fare: ",Datos['Fare'].isin([0]).sum())
```

```python
#Remplazar todos los ceros con el valor medio (promedio) de la columna.

Datos['Fare']=Datos['Fare'].replace(0,Datos['Fare'].mean())
```

```python
# Revisión nuevamente de datos Faltantes

print("Cantidad de registros con cero en la columna Fare: ",Datos['Fare'].isin([0]).sum())
```

```python
# Revisión de datos Faltantes

print("Cantidad de registros con datos faltantes en la columna Age: ",Datos['Age'].isna().sum())
```

```python
#Remplazar todos los valores faltantes con el valor medio (promedio) de la columna.

Datos["Age"].fillna(Datos["Age"].mean(),inplace= True)
```

```python
# Revisión nuevamente de datos Faltantes
    
print("Cantidad de registros con cero en la columna Age: ",Datos['Age'].isin([]).sum())
```

```python
# Eliminación de columna Cabin

Datos.drop('Cabin', axis=1, inplace=True)
```

```python
# Revisión de datos Faltantes

print("Cantidad de registros con datos faltantes en la columna Embarked: ",Datos['Embarked'].isna().sum())
```

```python
Datos.dropna(inplace=True)
```

```python
Datos.info()
```

```python
sex = pd.get_dummies(Datos['Sex'],drop_first=True)
embark = pd.get_dummies(Datos['Embarked'],drop_first=True)
```

```python
Datos.head()
```

```python
Datos.drop(['Sex','Embarked','Name','Ticket'],axis=1, inplace=True)
```

```python
Datos = pd.concat([Datos,sex,embark],axis=1)
```

```python
Datos.head()
```

```python
Datos.info()
```

```python
# Total de personas sobrevivientes

Datos['Survived'].value_counts()
```

```python
Datos[['Survived']].head(40)
```

```python
# 4.Dividir el dataset en Train y Test para evaluar correctamente el modelo.
```

```python
# Porcentaje de datos para entrenamiento (train) = 70%, y para prueba (test) = 30%
# Se crean vaiables train y test para almacenar la información del almacenamiento.
X = Datos.drop('Survived',axis=1) # se cargan los datos de train, sin la columna Survived
Y = Datos['Survived']# se cargan los datos del test, con la columna Survived

# Dividir el dataset en dos columnas entrenamiento y prueba (train y test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=0)
```

```python
print (X_train)
```

```python
print (Y_test)
```

```python
# 5.Entrenar el modelo configurando los diferentes hiperparámetros.
```

```python
# Construir el modelo de regresión logistica
modelo_regresion = LogisticRegression(solver = "liblinear")
modelo_regresion.fit (X_train , Y_train)

# crea el modelo de predicción
prediccion = modelo_regresion.predict(X_test) #predecirnlos valores apartir del conjunto de datosde validación
```

```python
print (Y_test)
```

```python
print (prediccion)
```

```python
# Desempeño de modelo de regresión

print(classification_report(Y_test,prediccion))
```

```python
# 7.Realizar las diferentes gráficas que permitan visualizar los resultados del modelo.
```

```python
cm=confusion_matrix(Y_test,prediccion)
disp=metrics.ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
# Curva de precisión del modelo
from sklearn.metrics import RocCurveDisplay
RocCurveDisplay.from_predictions(Y_test, prediccion)
plt.show()
```
